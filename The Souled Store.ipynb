```python
product_id=[]
main_title=[]
brand=[]
collaboration=[]
category=[]
color=[]
sizes=[]
fittype=[]
actualPrice=[]
currentPrice=[]
status=[]


count=1
for product_detail in product_details_json:
    product = json.loads(product_detail)
    
    for size_json in product["variant"]:
        
        product_id.append(f"TSS{count}")
        main_title.append(product["product"])
        brand.append("The Souled Store")
        collaboration.append(product["firstname"])
        category.append(product["category"])
        
        try:
            color.append(size_json["attributes"][1]["value"])
        except:
            color.append(np.nan)
         
        try:
            sizes.append(size_json["attributes"][0]["value"])
        except:
            sizes.append(np.nan)
        fittype.append(product["fittype"])
        actualPrice.append(product['mrp'])
        currentPrice.append(product["price"])
        
        if size_json["stock"] != 0:
            status.append("AVAILABLLE")
        else:
            status.append("OUT_OF_STOCK")
    
    count+=1
```

```python
df =pd.DataFrame(data={
        'product_id':product_id,
        'title':main_title,
        'brand':brand,
        'collaboration':collaboration,
        'category':category,
        'color':color,
        'size':sizes,
        'fittype':fittype,
        'actual_price':actualPrice,
        'current_price':currentPrice,
        'status':status,
    })
```
