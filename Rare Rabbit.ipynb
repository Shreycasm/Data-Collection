```python
import pandas as pd
import numpy as np
import seaborn as sns
import requests
import json
from bs4 import BeautifulSoup
```

```python
links= {'T-shirts': {'url_end': 'rr-tshirts-all'},
 'Polo Logo': {'url_end': 'pare-and-pareto-collection'},
 'Polo':{'url_end':'rr-men-polo-t-shirts'},        
 'Sweater': {'url_end': 'rare-rabbit-sweaters'},
 'Sweatshirt': {'url_end': 'sweatshirts-for-men'}}
```

```python


url = "https://1.search.argoid.com/rarerabbit/v1/search/query"

for cat, link in links.items():
    payload = json.dumps({
      "instanceId": "rarerabbit",
      "query": link['url_end'],
      "appSource": "WEB",
      "searchId": "",
      "userIds": {
        "anonymousUserId": "",
        "registeredUserId": ""
      },
      "state": 9,
      "page": 0,
      "pageSize": 650,
      "requestType": "PLP",
      "requestAttributes": "",
      "resultSize": 0,
      "sortField": "",
      "isSortAscending": True
    })
    headers = {
      'Accept': '*/*',
      'Accept-Language': 'en-US,en;q=0.8',
      'Authorization': 'Basic c2VhcmNoQHFhbGFyYS5jb206cWFsYXJhQDEyMw==',
      'Connection': 'keep-alive',
      'Content-Type': 'application/json',
      'Origin': 'https://thehouseofrare.com',
      'Referer': 'https://thehouseofrare.com/',
      'Sec-Fetch-Dest': 'empty',
      'Sec-Fetch-Mode': 'cors',
      'Sec-Fetch-Site': 'cross-site',
      'Sec-GPC': '1',
      'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/127.0.0.0 Safari/537.36',
      'sec-ch-ua': '"Not)A;Brand";v="99", "Brave";v="127", "Chromium";v="127"',
      'sec-ch-ua-mobile': '?0',
      'sec-ch-ua-platform': '"Windows"'
    }

    response = requests.request("POST", url, headers=headers, data=payload)
    
    links[cat]["content"] = response.json()


```

```python
product_id=[]
main_title=[]
brand=[]
category=[]
color=[]
sizes=[]
actualPrice=[]
currentPrice=[]
status=[]
imageurl=[]

count=1
for cat, link in links.items():
    for num, item in enumerate(link["content"]["assets"]):
        for size, status_ in json.loads(item["record"]["relatedVariants"]).items():

            product_id.append(f'RR{count}')
            main_title.append(item["record"]["main_title"])
            brand.append(item["record"]["brand"])
            category.append(item["record"]["type"])
            color.append(item["record"]["color"])
            sizes.append(size)
            actualPrice.append(item["record"]["actualPrice"])
            currentPrice.append(item["record"]["currentPrice"])
            status.append(item["record"]['status'])
            imageurl.append(item["record"]["imageurl"])
            
        count+=1
```

```python
df = (
    pd.DataFrame(data={
        'product_id':product_id,
        'title':main_title,
        'brand':brand,
        'category':category,
        'color':color,
        'size':sizes,
        'actual_price':actualPrice,
        'current_price':currentPrice,
        'status':status,
        'imageurl':imageurl
    })
)
```
